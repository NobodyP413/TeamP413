```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

traffic_data = pd.read_csv('traffic_light_dataTable.csv') #opening dataset
input_data = traffic_data.drop(columns = ['TL_A1', 'TL_A2', 'TL_B']) # getting the sample input data from dataset
output_data = traffic_data.drop(columns = ['Lane_A1', 'Lane_A2', 'Lane_B']) # getting the sample out data from dataset

#creating model
model = DecisionTreeClassifier()
# feeding the model of the data that we get from dataset
model.fit(input_data, output_data)

#getting input from the user that he want to predict
lane_a1 = int(input(f'Traffic status on Lane_A1: '))
lane_a2 = int(input(f'Traffic status on Lane_A2: '))
lane_b = int(input(f'Traffic status on Lane_B: '))


#getting the prediction of the model
predictions = model.predict([[lane_a1,lane_a2,lane_b]])

green_light_a1 = int(predictions[0][0])
green_light_a2 = int(predictions[0][1])
green_light_b = int(predictions[0][2])

############## print out purpose only don't be afraid we got this #####################
yellow_light_a1 = False
print(f'Traffic Light A1: Go \t Traffic Light A2: Go \t Traffic Light B: Stop')
while green_light_a1 != 0:
    print(f'\t{green_light_a1}\t\t\t{green_light_a2}')
    if green_light_a1 == 1 and not yellow_light_a1:
        green_light_a1 = 6
        yellow_light_a1 = True
    green_light_a1 -= 1
    green_light_a2 -= 1
    
yellow_light_a2 = False
print(f'Traffic Light A1: Stop \t Traffic Light A2: Go \t Traffic Light B: Go')
while green_light_a2 != 0:
    print(f'\t\t\t\t{green_light_a2}\t\t\t{green_light_b}')
    if green_light_a2 == 1 and not yellow_light_a2:
        green_light_a2 = 6
        green_light_b = 6
        yellow_light_a2 = True
    green_light_a2 -= 1
    green_light_b -= 1
    
```
